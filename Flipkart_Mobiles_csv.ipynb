# **IMPORT LIBRARY**

```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

sns.set(style="whitegrid", palette="viridis")  # Premium aesthetic palette

df = pd.read_csv("Flipkart_Mobiles.csv")
df.head()
```

 **DESCRIBE-ye sab graph ka base setup hai**

```python
df.describe()
```

# **Brand Count Plot**
Kaunsa brand sabse zyada listed hai wo batata hai

```python
plt.figure(figsize=(10,5))
sns.countplot(data=df, x="Brand", order=df['Brand'].value_counts().index)
plt.title("Number of Mobiles by Brand")
plt.xticks(rotation=45)
plt.show()
```

```python
plt.figure(figsize=(12,6))
sns.lineplot(data=df, x="Selling Price", y="Rating", hue="Brand", linewidth=2.5)
plt.title("Brand-wise Rating vs Selling Price", fontsize=16, weight="bold")
plt.show()
```

Line chart check karta hai kis price range me kis brand ki rating high hai

hue="Brand" se colour-coded brand comparison milta hai.

```python
plt.figure(figsize=(12,6))
sns.barplot(data=df, x="Brand", color="green", y="Selling Price", estimator="mean")
plt.title("Average Selling Price per Brand", fontsize=16, weight="bold")
plt.xticks(rotation=45)
plt.show()
```

# **Color Distribution**
Kaunsa color sabse zyada bikta hai.

```python
plt.figure(figsize=(10,5))
sns.countplot(data=df, x="Color", order=df['Color'].value_counts().index[:10])
plt.title("Color Distribution")
plt.xticks(rotation=45)
plt.show()
```

# **MEMORY VS COUNT**
4GB/6GB/8GB variant ka distribution

```python
plt.figure(figsize=(8,5))
sns.countplot(data=df, x="Memory")
plt.title("Memory Variant Distribution")
plt.show()
```

# **. Rating Distribution (Histogram)**
Phones ki rating overall kitni achi hai

```python
plt.figure(figsize=(8,5))
sns.histplot(df['Rating'], kde=True, color="darkblue")
plt.title("Rating Distribution")
plt.show()
```

# **BOX PLOT -BRAND WISE PRICE SPEAD**

```python
plt.figure(figsize=(12,6))
sns.boxplot(data=df, x="Brand", y="Selling Price", color = 'red')
plt.title("Price Variation per Brand", fontsize=16, weight="bold")
plt.xticks(rotation=45)
plt.show()
```

VIOLIN PLOT RATING DISTRIBUSTION BY BRAND

```python
plt.figure(figsize=(12,6))
sns.violinplot(
    data=df,
    x="Brand",
    y="Rating",
    inner="quartile",       # andar quartiles dikhata hai (premium look)
    linewidth=1.3           # smooth border
)
plt.title("Rating Distribution per Brand", fontsize=16, weight="bold")
plt.xticks(rotation=45)
plt.xlabel("Brand")
plt.ylabel("Rating")
plt.tight_layout()
plt.show()
```

# **SCATTER PLOT - RATTING VS ORGINAL PRICE**

```python
plt.figure(figsize=(12,6))
sns.scatterplot(data=df, x="Original Price", y="Rating", hue="Brand", size="Selling Price")
plt.title("Rating vs Original Price", fontsize=16, weight="bold")
plt.show()
```
