# **integrating the data set and plot our first graph**

```python
df=pd.read_csv('Cardio.csv')
df
```

```python
df.info()
```

```python
sns.set_style('whitegrid')
sns.set_palette('viridis')
sns.set_context('notebook')
sns.lineplot(x='Age',y='Usage',data=df,hue='Gender',marker='o')
plt.show()
```

```python
plt.figure(figsize=(150,200),dpi=300)
df_filtered = df[df['Gender'] != 'Female'].copy()
g = sns.catplot(x='Age',y='Usage',data=df_filtered,kind='bar',errorbar=None)
g.set_xticklabels(rotation=45, ha="right")
plt.tight_layout()
plt.show()
```

```python
g=sns.barplot(x='Age',y='Usage',data=df,hue='MaritalStatus',errorbar=None)
g.set_xticklabels(g.get_xticklabels(), rotation=45, ha="right")
```

```python
sns.pairplot(df)
plt.show
```

```python
numeric_df = df.select_dtypes(include=['number'])
numeric_df2=numeric_df.corr()
sns.heatmap(data=numeric_df2, annot=True, cmap='coolwarm')
plt.title('Correlation Heatmap of Numeric Columns')
plt.show()
```

```python
e=sns.stripplot(x='Age',y='Usage',data=df,hue='Gender',jitter=True)
e.set_xticklabels(e.get_xticklabels(), rotation=45, ha="right")
```

```python
sns.stripplot(x='Product',y='Age',data=df,hue='Gender',jitter=False)
```

```python
product_counts = df['Product'].value_counts()

plt.pie(x=product_counts.values, labels=product_counts.index, autopct='%1.1f%%')
plt.title('Distribution of Products')
plt.show()
```
